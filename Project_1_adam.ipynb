```python
#get events for Berlin

import requests

events_berlin = []

# Page 0
url_page_0 = (
    "https://app.ticketmaster.com/discovery/v2/events?"
    "apikey=RDUun7PpJbEjNNkxqNBPXd8XADKRARVZ"
    "&city=Berlin"
    "&keyword=music"
    "&size=200"
    "&startDateTime=2025-06-01T00:00:00Z"
    "&endDateTime=2025-12-31T23:59:59Z"
    "&page=0"
)
response = requests.get(url_page_0).json()
events_berlin += response.get('_embedded', {}).get('events', [])

# Page 1
url_page_1 = url_page_0.replace("page=0", "page=1")
response = requests.get(url_page_1).json()
events_berlin += response.get('_embedded', {}).get('events', [])

len(events_berlin)
```

```python
df_berlin['city'] = 'Berlin'
df_toulouse['city'] = 'Toulouse'

# Merge both DataFrames
df_all = pd.concat([df_berlin, df_toulouse], ignore_index=True)

print(df_all.shape)
df_all.head()
```

```python
#clean database
import pandas as pd

df_clean = df_all.copy()

df_clean['start_datetime'] = df_clean['dates'].apply(lambda x: x.get('start', {}).get('dateTime') if isinstance(x, dict) else None)
df_clean['status'] = df_clean['dates'].apply(lambda x: x.get('status', {}).get('code') if isinstance(x, dict) else None)

df_clean['genre'] = df_clean['classifications'].apply(
    lambda x: x[0].get('genre', {}).get('name') if isinstance(x, list) and len(x) > 0 else None
)
df_clean['segment'] = df_clean['classifications'].apply(
    lambda x: x[0].get('segment', {}).get('name') if isinstance(x, list) and len(x) > 0 else None
)

df_clean['city'] = df_clean['_embedded'].apply(
    lambda x: x.get('venues', [{}])[0].get('city', {}).get('name') if isinstance(x, dict) else None
)


columns_to_drop = ['images', 'sales', 'dates', 'classifications', 'promoter', 'promoters', 'seatmap', 'ticketing', '_links', '_embedded', 'outlets', 'description', 'locale', 'test', 'type']
df_clean.drop(columns=[col for col in columns_to_drop if col in df_clean.columns], inplace=True)

# Show cleaned dataframe
print(df_clean.head())
```

```python
grouped = df_clean.groupby(['city', 'genre']).size().reset_index(name='count')

# Step 2: Calculate total events per city
city_totals = df_clean['city'].value_counts().to_dict()

# Step 3: Map city totals to each row
grouped['city_total'] = grouped['city'].map(city_totals)

# Step 4: Calculate percentage per genre within each city
grouped['percentage'] = (grouped['count'] / grouped['city_total']) * 100
grouped['percentage'] = grouped['percentage'].round(2)
```
